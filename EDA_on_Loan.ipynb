#### We can bin the Income columns into Very Low Class, Lower class, Middle class, High class and Rich class based on theier salary 

This Can be done with the help of qcut and quantiles.We can assume that Salary between 0 - 20 percent quantile as Very low , 20-45 percentile as Low , 45-75 as Middle class , 75 - 95 as High Middle class and 95 above as Rich class

```python
app_data['AMT_INCOME_TOTAL_BINS']= pd.qcut(app_data.AMT_INCOME_TOTAL, q=[0, 0.2, 0.45, 0.75, 0.95, 1],
                                                                labels=['Very_low_class', 'Low_class', 'Middle_class','High_Middle','Rich_class'])

```

```python
app_data['AMT_INCOME_TOTAL_BINS'].value_counts(normalize=True).plot.barh()
plt.title('Income distribution based on Class')
plt.show()
```

```python
mer_data = mer_data.replace('XNA',np.nan)
```

```python
def pivot_table_heatmap(Var_1, Var_2):
    # Create the pivot table
    table = pd.pivot_table(mer_data, values='TARGET', index=[Var_1], columns=[Var_2], aggfunc=np.mean)
    
    # Create the heatmap
    plt.figure(figsize=(10, 5))
    sns.heatmap(table, annot=True, cmap="YlGnBu", cbar_kws={'label': '% of Loan-Payment Difficulties'})
    plt.title(f'{Var_1} vs {Var_2} on TARGET', fontdict={'fontsize': 20})
    plt.ylabel(Var_1)
    plt.xlabel(Var_2)
    plt.show()
```

We can infer that most of the application Canceled and refused belongs to Male

From this we can conclude that Males have more default rate than females

We can infer that the Low skill laborers tend to default more

We can visualize that the percentage of Defaulters is high for Civil marriage and Single/not married

People living in 'Rented Apartments' with a previous application type of 'unused offer' default more frequently than other groups. In all other categories, individuals who live with their parents and had a previous application that was 'cancelled' or 'refused' are more likely to default
